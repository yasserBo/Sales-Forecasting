```python
import pandas as pd

file_path = 'retail1.csv'
data = pd.read_csv(file_path, encoding='latin1')  # Common encoding for non-UTF-8 files
print(data.head())
print(data.info())
```

```python
# Drop rows where Customer ID is missing
data = data.dropna(subset=['Customer ID'])

# Replace missing Descriptions with 'Unknown'
data['Description'] = data['Description'].fillna('Unknown')

# Verify changes
print(data.info())
```

```python
# Convert InvoiceDate to datetime format
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], errors='coerce')

# Check for invalid dates
print("Invalid Dates:", data['InvoiceDate'].isnull().sum())

# Confirm changes
print(data.info())
```

```python
# Remove duplicate rows
data = data.drop_duplicates()

# Confirm no duplicates remain
print(f"Total duplicates remaining: {data.duplicated().sum()}")
```

```python
# Summary statistics
print(data[['Quantity', 'Price']].describe())

# Check for negative values
print("Negative Quantity:", (data['Quantity'] < 0).sum())
print("Negative Price:", (data['Price'] < 0).sum())
```

```python
# Remove transactions with invalid values
data = data[(data['Quantity'] > 0) & (data['Price'] > 0)]

# Verify
print(data.info())
```

```python
# Save the cleaned dataset
data.to_csv('cleaned_sales_data.csv', index=False)
print("Cleaned dataset saved successfully!")
```

```python
import matplotlib.pyplot as plt

# Add a 'Revenue' column
data['Revenue'] = data['Quantity'] * data['Price']

# Extract Year-Month for grouping
data['YearMonth'] = data['InvoiceDate'].dt.to_period('M')

# Group data and calculate revenue
monthly_sales = data.groupby('YearMonth')['Revenue'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

# Plot sales trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Revenue'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month-Year')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Group by Description and aggregate Quantity and Revenue
product_sales = data.groupby('Description').agg({
    'Quantity': 'sum',
    'Revenue': 'sum'
}).sort_values(by='Revenue', ascending=False)  # You can also sort by Quantity to see physical sales volume

# Display top 10 products by revenue
top_products_by_revenue = product_sales.head(10)
print("Top 10 Products by Revenue:")
print(top_products_by_revenue)

# Display top 10 products by quantity sold
top_products_by_quantity = product_sales.sort_values(by='Quantity', ascending=False).head(10)
print("\nTop 10 Products by Quantity Sold:")
print(top_products_by_quantity)

# Plot top products by revenue
top_products_by_revenue['Revenue'].plot(kind='bar', title='Top 10 Products by Revenue')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha="right")
plt.show()

# Plot top products by quantity
top_products_by_quantity['Quantity'].plot(kind='bar', color='green', title='Top 10 Products by Quantity Sold')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Calculate total revenue per customer
customer_revenue = data.groupby('Customer ID')['Revenue'].sum().sort_values(ascending=False)

# Display top 10 high-value customers
print("Top 10 High-Value Customers:")
print(customer_revenue.head(10))

# Plot top 10 high-value customers
customer_revenue.head(10).plot(kind='bar', title='Top 10 High-Value Customers')
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Count the number of invoices per customer
customer_frequency = data.groupby('Customer ID')['Invoice'].nunique()

# Display the most frequent buyers
print("Top 10 Frequent Buyers:")
print(customer_frequency.sort_values(ascending=False).head(10))

# Plot purchase frequency of top 10 customers
customer_frequency.sort_values(ascending=False).head(10).plot(kind='bar', color='purple', title='Top 10 Frequent Buyers')
plt.xlabel('Customer ID')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Define spending categories based on revenue quantiles
spending_tiers = pd.qcut(customer_revenue, q=[0, 0.25, 0.5, 0.75, 1], labels=['Low', 'Medium', 'High', 'Very High'])

# Display the distribution of spending categories
print("Customer Spending Tiers Distribution:")
print(spending_tiers.value_counts())

# Pie chart of spending tiers
spending_tiers.value_counts().plot(kind='pie', title='Customer Spending Tiers', autopct='%1.1f%%')
plt.ylabel('')  # Remove the 'Spending Tiers' label on the y-axis
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sum revenue by country
sales_by_country = data.groupby('Country')['Revenue'].sum().sort_values(ascending=False)

# Display top 10 countries by revenue
print("Top 10 Countries by Revenue:")
print(sales_by_country.head(10))

# Plot sales by country
plt.figure(figsize=(10, 6))
sales_by_country.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Countries by Revenue')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract month from the InvoiceDate
data['Month'] = data['InvoiceDate'].dt.month

# Group by month and sum up the revenue
seasonal_trends = data.groupby('Month')['Revenue'].sum()

# Plot the seasonal sales trend
plt.figure(figsize=(10, 6))
seasonal_trends.plot(kind='line', marker='o', color='darkblue')
plt.title('Seasonal Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Assuming 'monthly_sales' is your DataFrame and it has a 'Revenue' column.
result = adfuller(monthly_sales['Revenue'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Check if the series is stationary
if result[1] > 0.05:
    print("Series is not stationary")
else:
    print("Series is stationary")
```

```python
# First differencing
monthly_sales['Revenue_diff'] = monthly_sales['Revenue'].diff().dropna()

# ADF test on the differenced series
result_diff = adfuller(monthly_sales['Revenue_diff'].dropna())  # Drop NA values that result from differencing
print('ADF Statistic:', result_diff[0])
print('p-value:', result_diff[1])

# Check if the series is now stationary
if result_diff[1] > 0.05:
    print("Series is still not stationary")
else:
    print("Series is now stationary")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Ensure numpy and statsmodels are imported correctly

# Simulating a sample time series data for 'Revenue'
monthly_sales = pd.DataFrame({
    'Revenue': np.random.randn(100).cumsum() + 100  # Random cumulative revenue data
})

# Differencing the data to make it stationary (change in Revenue)
monthly_sales['Revenue_diff'] = monthly_sales['Revenue'].diff().dropna()

# Check the length of the differenced series and adjust lags accordingly
available_lags = len(monthly_sales['Revenue_diff'].dropna())
lags = min(20, available_lags - 1)  # Setting lags to a maximum of 20 or less if not enough data

# Plot ACF and PACF
fig, ax = plt.subplots(1, 2, figsize=(16, 3))
plot_acf(monthly_sales['Revenue_diff'].dropna(), lags=lags, ax=ax[0])  # ACF plot
plot_pacf(monthly_sales['Revenue_diff'].dropna(), lags=lags, ax=ax[1])  # PACF plot
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Model setup (using p=1, d=1, q=1 as a starting point)
model = ARIMA(monthly_sales['Revenue'], order=(1,1,1))
model_fit = model.fit()

# Print out the summary of the model
print(model_fit.summary())

# Forecasting future values
forecast = model_fit.forecast(steps=12)

# Plot the historical data and the forecast
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales['Revenue'], label='Historical')
plt.plot(pd.date_range(monthly_sales.index[-1], periods=12, freq='M'), forecast, label='Forecast')
plt.title('Future Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import stats

# Example data: Sales before and after a specific campaign
sales_before = np.array([100, 105, 98, 97, 103, 101, 106, 104])
sales_after = np.array([110, 108, 115, 120, 125, 130, 118, 112])

# Conduct a t-test
t_stat, p_value = stats.ttest_ind(sales_after, sales_before, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant increase in sales after the campaign.")
else:
    print("Fail to reject the null hypothesis: There is no significant change in sales after the campaign.")
```

```python
import pandas as pd

# Load the cleaned data 
data = pd.read_csv('cleaned_sales_data.csv')

# Verify the DataFrame and its columns
print(data.columns)
print(data.head())

```

```python
# Ensure that Quantity and Price columns exist
if 'Quantity' in data.columns and 'Price' in data.columns:
    data['Revenue'] = data['Quantity'] * data['Price']
else:
    print("Error: Quantity and/or Price columns missing from the data.")
```

```python
# Save the updated dataset with the 'Revenue' column
data.to_csv('updated_cleaned_sales_data.csv', index=False)
print("Updated dataset with Revenue saved successfully!")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Selecting features for clustering
features = data[['Quantity', 'Price', 'Revenue']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Applying K-means Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(features_scaled)
data['Product_Cluster'] = clusters

# Analyzing the results
cluster_summary = data.groupby('Product_Cluster').agg({
    'Quantity': 'mean',
    'Price': 'mean',
    'Revenue': 'mean'
}).round(2)

print(cluster_summary)
```
